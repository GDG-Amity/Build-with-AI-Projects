```python
#Import Important Libraries
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('adult.csv')
```

```python
df=df[['workclass','education','marital-status','occupation','race','sex','Salary']]
```

```python
df.head()
```

### Data contains both Ordinal Categorical Data (Education) as well as Nominal Categorical Data (race)

### Let us deal with Ordinal Data first 

### For independant variables we use Ordinal Encoder and if target variable is categorical, we will use Label Encoding

### Ordinal Encoding

```python
df['education'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Define the ordered categories for the 'education' feature
education_order = [
    ' Preschool', ' 1st-4th', ' 5th-6th', ' 7th-8th', ' 9th', ' 10th',
    ' 11th', ' 12th', ' HS-grad', ' Some-college', ' Assoc-voc',
    ' Assoc-acdm', ' Bachelors', ' Masters', ' Prof-school', ' Doctorate'
]

# Initialize the OrdinalEncoder with the specified order
oe = OrdinalEncoder(categories=[education_order])  # Note the double brackets

# Fit and transform the 'education' column
df['education_encoded'] = oe.fit_transform(df[['education']])
```

```python
df.head()
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['Salary_encoded']=le.fit_transform(df['Salary'])
```

```python
df.sample(10)
```

#### 3. One Hot Encoding using sklearn

```python
X=df[['race']]
Y=df['Salary_encoded']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(drop='first',sparse_output=False)
X_train_new = ohe.fit_transform(X_train)
X_test_new = ohe.fit_transform(X_test)
```
